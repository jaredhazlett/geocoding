```python
#geocoding practice
import requests
import json
from config import gkey
```

```python
lat_list = 40.714224
lng_list = -73.961452
target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
                'latlng={0},{1}&key={2}').format(lat_list, lng_list, gkey)
```

```python
geo_data = requests.get(target_url).json()
print(json.dumps(geo_data, indent=4, sort_keys=True))
```

```python
sublocality = geo_data["results"][0]["address_components"][3]["long_name"]
postal_code = geo_data["results"][0]["address_components"][7]["long_name"]

print(f'{sublocality}, {postal_code}')
```

```python
coordinates = [
    (40.71, -74.00),
    (30.26, -97.74),
    (46.87, -96.78),
    (47.60, -122.33),
    (32.71, -117.16)
]
```

```python
for i, tuple in enumerate(coordinates):
    element_one = tuple[0]
    element_two = tuple[1]
    target_url = (f'https://maps.googleapis.com/maps/api/geocode/json?'
                'latlng={0},{1}&key={2}').format(element_one, element_two, gkey)
    geo_data = requests.get(target_url).json()
    print(json.dumps(geo_data, indent=4, sort_keys=True))
```
