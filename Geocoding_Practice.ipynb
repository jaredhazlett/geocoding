```python
#geocoding practice
import requests
import json
from config import gkey
```

```python
lat_list = 40.714224
lng_list = -73.961452
target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
                'latlng={0},{1}&key={2}').format(lat_list, lng_list, gkey)
```

```python
geo_data = requests.get(target_url).json()
print(json.dumps(geo_data, indent=3, sort_keys=True))
```

```python
address = geo_data["results"][0]["formatted_address"]
print(address)
```
